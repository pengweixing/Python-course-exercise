```python
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt
```

```python
poisson = stats.poisson(10)
```

```python
samples = poisson.rvs(size=1000)
```

```python
x = np.arange(0, 20)
pmf = poisson.pmf(x)
plt.bar(x, pmf)
plt.title('PMF of Poisson Distribution')
plt.show()
```

```python
cdf = poisson.cdf(x)
plt.step(x, cdf)
plt.title('CDF of Poisson Distribution')
plt.show()
```

```python
plt.hist(samples, bins=15, density=True, label='Histogram')
plt.ylabel('Frequency')
plt.title('Histogram of Random Samples')
plt.legend()
plt.show()
```

```python
normal = stats.norm(1, 1)
```

```python
samples = normal.rvs(size=1000)
```

```python
x = np.linspace(-4, 4, 100)
pdf = normal.pdf(x)
plt.plot(x, pdf)
plt.xlabel('x')
plt.ylabel('f(x)')
plt.title('PDF of Normal Distribution')
plt.show()
```

```python
cdf = normal.cdf(x)
plt.plot(x, cdf)
plt.xlabel('x')
plt.ylabel('F(x)')
plt.title('CDF of Normal Distribution')
plt.show()
```

```python
plt.hist(samples, bins=20, density=True, label='Histogram')
plt.ylabel('Frequency')
plt.title('Histogram of Random Samples')
plt.legend()
plt.show()
```

```python
data1 = stats.norm.rvs(loc=0, scale=1, size=100)
data2 = stats.norm.rvs(loc=1.5, scale=2.5, size=100)

# Perform Kolmogorov-Smirnov test
test_stat, p_value = stats.ks_2samp(data1, data2)

# Print results
print('Test statistic:', test_stat)
print('P-value:', p_value)
```

```python
# Interpret results
if p_value < 0.05:
    print('We reject the null hypothesis that the samples come from the same distribution.')
else:
    print('We fail to reject the null hypothesis that the samples come from the same distribution.')
```
